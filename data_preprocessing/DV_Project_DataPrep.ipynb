## DATA PREPROCESSING FOR RIDGECHART

```python
d3 = data[["Country", "Year", "BMI", "HIV/AIDS"]]
d3
```

```python
d3 = d3.sort_values(by=["Country", "Year"], ascending=[True, True])
d3
```

```python
if d3.isna().any().any():
    print("Il dataset contiene valori NaN.")
else:
    print("Il dataset non contiene valori NaN.")
```

```python
# Seleziona solo le righe che contengono valori NaN
valori_nan = d3.loc[d3.isna().any(axis=1)]

# Stampa i valori NaN
print("Valori NaN nel DataFrame:")
print(valori_nan)
```

```python
d3 = d3.dropna()
d3
```

```python
# Funzione per trasformare i dati
def transform_data(data):
    # Seleziona le colonne necessarie
    data = data[['Country', 'Year', 'BMI', 'HIV/AIDS']]

    # Melt i dati per BMI
    bmi_data = data.pivot(index='Country', columns='Year', values='BMI').reset_index()
    bmi_data['Type'] = 'BMI'
    
    # Melt i dati per HIV/AIDS
    hiv_data = data.pivot(index='Country', columns='Year', values='HIV/AIDS').reset_index()
    hiv_data['Type'] = 'HIV/AIDS'
    
    # Unisci i due dataframe
    transformed_data = pd.concat([bmi_data, hiv_data], ignore_index=True)
    
    return transformed_data
```

```python
transformed_data = transform_data(d3)
transformed_data
```

```python
transformed_data = transformed_data.sort_values(by=['Country', 'Type']).reset_index(drop=True)
transformed_data
```

```python
numeric_data = transformed_data.iloc[:, 1:-1]
```

```python
min_value = numeric_data.min().min()
min_value
```

```python
max_value = numeric_data.max().max()
max_value
```

```python
transformed_data.to_csv("Ridgechart_Data.csv", index=False)
```

## DATA PREPROCESSING FOR HEATMAP

```python
d4 = data[["Country", "Year", "Adult_Mortality", "Infant Deaths", "BMI", "HIV/AIDS", "GDP"]]
d4
```

```python
d4.rename(columns={'Infant Deaths': 'Infant_Deaths'}, inplace=True)
d4
```

```python
if d4.isna().any().any():
    print("Il dataset contiene valori NaN.")
else:
    print("Il dataset non contiene valori NaN.")
```

```python
# Seleziona solo le righe che contengono valori NaN
valori_nan = d4.loc[d4.isna().any(axis=1)]

# Stampa i valori NaN
print("Valori NaN nel DataFrame:")
print(valori_nan)
```

```python
d4 = d4.dropna()
d4
```

```python
variables = ["Adult_Mortality", "Infant_Deaths", "BMI", "HIV/AIDS", "GDP"]
correlation_dict = {}
for country in d4['Country'].unique():
    country_data = d4[d4['Country'] == country]
    country_data = country_data[variables]  # Consider only the variables of interest
    correlation_matrix = country_data.corr()  # Calculate correlation matrix
    correlation_dict[country] = correlation_matrix
```

```python
for country, corr_matrix in correlation_dict.items():
    print(f"Correlazione per {country}:")
    print(corr_matrix)
    print("\n")
```

```python
# Creare una lista per raccogliere i dati
data_for_csv = []

# Loop through the dictionary and prepare the data
for country, corr_matrix in correlation_dict.items():
    # Melt the correlation matrix to long format
    melted_corr = corr_matrix.reset_index().melt(id_vars='index')
    melted_corr.columns = ['Variable1', 'Variable2', 'Correlation']
    melted_corr['Country'] = country
    data_for_csv.append(melted_corr)

# Concatenate all the dataframes
final_df = pd.concat(data_for_csv, ignore_index=True)
final_df

# Salvare il dataframe concatenato in un CSV
#final_df.to_csv('correlation_matrices.csv', index=False)
```

```python
col = 'Country'
final_df = final_df[[col] + [c for c in final_df if c != col]]
final_df
```

```python
final_df['Correlation'] = final_df['Correlation'].round(2)
final_df
```

```python
final_df.to_csv('correlation_matrices.csv', index=False)
```
